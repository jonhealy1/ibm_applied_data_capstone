#### Import re and creats a function clean the text

```python
import re

TAG_RE = re.compile(r'<[^>]+>')

def remove_tags(text):
    return TAG_RE.sub('', text)
```

#### Get data from wikipedia and make lists of all the data

```python
# import requests
# from bs4 import BeautifulSoup
Postcode = []
Neigh = []
Borough = []
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'lxml')
for items in soup.find('table', class_='wikitable').find_all('tr')[1::1]:
    data = items.find_all(['th','td'])
    if (data[1].text != "Not assigned"):
        try:
            title = remove_tags(data[0].text)
            bor = remove_tags(data[1].text)
            code = remove_tags(data[2].text)
            code = code.strip('\n')
        #line = line.replace('\n','')
        except IndexError:pass
    # print("{}".format(title))
        Postcode.append(title)
        Neigh.append(code)
        Borough.append(bor)
```

#### Add data to dataframe

```python
import pandas as pd
df2 = pd.DataFrame()

df2["PostalCode"] = Postcode
df2["Borough"] = Borough
df2["Neighborhood"] = Neigh
```

#### Group the data by postal code

```python
df2 = df2.groupby(['PostalCode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
```
